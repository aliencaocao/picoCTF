```python
from functools import reduce

def phase1(x: list, y: int) -> list:
    #print(f'Enter phase1, x: {x}, y: {y}')
    if x:
        #print(f'IF x: x: {x}')
        if x[(-2)] == y:
            print(f'IF x[(-2)] == y: x[(-2)]: {x[(-2)]}, y: {y}')
            if x[(-1)] < 255:
                print(f'IF x[(-1)] < 255: x[(-1)]: {x[(-1)]}')
                x[(-1)] += 1
        #print(f'Returning x: {x}')
        return x
    else:
        print(f'ELSE: x: {x}')
        x.extend([y, 1])
        print(f'x extended with [y, 1]: {[y, 1]}')
        print(f'Returning x: {x}')
        return x


def phase2(x: int) -> int:
    print('Enter phase2, x:', x)
    print(f'Returning: {~x & 255}')
    return ~x & 255


with open('flag.jpg', 'rb') as (f):
    data = f.read()
encoded = map(phase2, reduce(phase1, data, []))
# with open('fakeflag.jpg.encoded', 'wb') as (f):
#     f.write(bytes(encoded))
```
